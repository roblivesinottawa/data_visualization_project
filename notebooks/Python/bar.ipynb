```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (10, 5))
ax = fig.add_axes([0, 0, 1, 1])
prog_languages = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 34, 45, 56, 67]
ax.bar(prog_languages, students)
plt.show()
```

```python
import numpy as np
```

```python
data = [[30, 20, 10, 40, 50],
        [10, 20, 30, 40, 50],
        [40, 30, 20, 10, 50]]

X = np.arange(5)
fig = plt.figure(figsize = (10, 5))
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)

plt.show()
```

```python
plt.figure(figsize = (10, 5))
countries = ['USA', 'Brazil', 'Russia', 'Spain', 'France', 'India']
total_deaths = [769000, 612000, 257000, 87810, 119000, 465000]
plt.bar(countries, total_deaths)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
countries = ['USA', 'Brazil', 'Russia', 'Spain', 'France', 'India']
total_deaths = [769000, 612000, 257000, 87810, 119000, 465000]
plt.bar(countries, total_deaths, width=0.9, align='center', color='black', edgecolor='black')
i = 1.0
j = 2000
for i in range(len(countries)):
    plt.annotate(total_deaths[i], (-0.1 + i, total_deaths[i] + j))

plt.legend(['Total Deaths'])
plt.title("Total Deaths by COVID-19 per Country")
plt.xlabel('Countries')
plt.ylabel('Total Deaths')
plt.show()
```
