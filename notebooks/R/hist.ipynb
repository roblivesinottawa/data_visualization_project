```r
library(ggplot2)
```

```r
library(ggplot2movies)
```

```r
df <- movies <- movies[sample(nrow(movies), 1000), ]
```

```r
head(df)
```

QPLOT

```r
qplot(rating, data = df, geom = "histogram", binwidth = 0.1, alpha = 0.8)
```

GGPLOT

```r
print(pl <- ggplot(df, aes(x = rating)) + geom_histogram())
```

ADDING COLOR

```r
print(pl <- ggplot(df, aes(x = rating)) + geom_histogram(binwidth = 0.1, color = "red", fill = "pink"))
```

ADDING LABELS

```r
pl <- ggplot(df, aes(x = rating)) + geom_histogram(binwidth = 0.1, color = "blue", fill = "black") + xlab("Movie Ratings") + ylab("Occurences") + ggtitle("Movie Ratings")
```

```r
print(pl)
```

CHANGE ALPHA (TRANSPARENCY)

```r
pl <- ggplot(df, aes(x = rating)) + geom_histogram(binwidth = 0.1, fill = "blue", alpha = 0.4) + xlab("Movie Ratings") + ylab("Occurences")
```

```r
print(pl)
```

LINETYPES

```r
pl <- ggplot(df, aes(x = rating)) + geom_histogram(binwidth = 0.1, color = "blue", fill = "pink", linetype = "dotted") + xlab("Movie Ratings") + ylab("Occurences") + ggtitle("Movie Ratings")
```

```r
print(pl)
```

ADVANCED AESTHETICS

```r
pl <- ggplot(df, aes(x = rating)) + geom_histogram(binwidth = 0.1, aes(fill = ..count.. )) + xlab("Movie Ratings") + ylab("Occurences")
print(pl)
```

```r
pl <- ggplot(df, aes(x = rating)) + geom_histogram(binwidth = 0.1, aes(fill = ..count.. )) + xlab("Movie Ratings") + ylab("Occurences") + scale_fill_gradient("Count", low = "blue", high = "red")
print(pl)
```

```r
pl + scale_fill_gradient("Count", low = "darkgreen", high = "lightblue")
```

```r
pl <- ggplot(df, aes(x = rating)) + geom_histogram(binwidth = 0.5, aes(y = ..density.. )) + geom_density(color = "red")
print(pl)
```
