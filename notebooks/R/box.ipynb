BOXPLOTS ARE A GOOD DAY TO DEPICT GROUPS OF NUMERICAL DATA THROUGH THEIR QUARTILES AND MEDIANS.

```r
library(ggplot2)
```

```r
df <- mtcars
print(head(df))
```

QPLOT

```r
pl <- qplot(factor(cyl), mpg, data=df, geom="boxplot")
print(pl)
```

GGPLOT

```r
pl <- ggplot(mtcars, aes(factor(cyl), mpg)) + geom_boxplot()
print(pl)
```
