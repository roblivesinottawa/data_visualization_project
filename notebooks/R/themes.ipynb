THEMES

```r
library(ggplot2)
```

```r
df <- mtcars
print(df)
```

```r
pl <- ggplot(df, aes(x=wt, y=mpg)) + geom_point()
print(pl)
```

```r
# theme_set(theme_minimal())
```

```r
pl <- ggplot(df, aes(x=wt, y=mpg)) + geom_point() + theme_dark()
print(pl)
```

```r
library(ggthemes)
```

```r
pl <- ggplot(df, aes(x=wt, y=mpg)) + geom_point() + theme_economist()
print(pl)
```

```r
pl <- ggplot(df, aes(x=wt, y=mpg)) + geom_point() + theme_fivethirtyeight()
print(pl)
```

```r
pl <- ggplot(df, aes(x=wt, y=mpg)) + geom_point() + theme_wsj()
print(pl)
```
