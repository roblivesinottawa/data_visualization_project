SCATTERPLOTS

```r
library(ggplot2)
```

```r
df <- mtcars
```

```r
print(head(df))
```

QPLOT

```r
print(qplot(wt, mpg, data = df))
```

ADDING A THIRD FEATURE

```r
print(qplot(wt, mpg, data = df, color = cyl))
```

```r
print(qplot(wt, mpg, data = df, size = cyl))
```

```r
print(qplot(wt, mpg, data = df, size = cyl, color = cyl))
```

SHOW 4 FEATURES

```r
print(qplot(wt, mpg, data = df, size = cyl, color = cyl, alpha = 0.6))
```

GGPLOT

```r
pl <- ggplot(data = df, aes(wt, mpg)) + geom_point()
print(pl)
```

ADDING THIRD FEATURE

```r
pl <- ggplot(data = df, aes(wt, mpg)) + geom_point(aes(color = cyl))
print(pl)
```

```r
pl <- ggplot(data = df, aes(x = wt, y =  mpg)) + geom_point(aes(color = factor(cyl)))
print(pl)
```

```r
pl <- ggplot(data = df, aes(x = wt, y =  mpg)) + geom_point(aes(size = factor(cyl)))
print(pl)
```

WITH SHAPES

```r
pl <- ggplot(data = df, aes(x = wt, y =  mpg)) + geom_point(aes(shape = factor(cyl)))
print(pl)
```

```r
pl <- ggplot(data = df, aes(x = wt, y =  mpg)) + geom_point(aes(shape = factor(cyl), color = factor(cyl)), size = 5, alpha = 0.6)
print(pl)
```

GRADIENT SCALES

```r
pl <- ggplot(data = df, aes(x = wt, y =  mpg)) + geom_point(aes(color = hp), size = 4) + scale_color_gradient(high = "red", low = "blue")

print(pl)
```
