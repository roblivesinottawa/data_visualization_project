BARPLOTS: THEY ARE BETTER USED WHEN DELING WITH CATEGORICAL DATA

```r
df <- mpg
```

```r
print(head(df))
```

```r
library(ggplot2)
```

```r
g <- ggplot(df, aes(class)) + geom_bar()
print(g)
```

```r
g <- ggplot(df, aes(class)) + geom_bar(aes(fill = drv))
print(g)
```

```r
g <- ggplot(df, aes(class)) + geom_bar(aes(fill = drv), position = "fill")
print(g)
```

```r
g <- ggplot(df, aes(class)) + geom_bar(aes(fill = drv), position = "dodge")
print(g)
```
