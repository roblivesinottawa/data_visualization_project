```r
library(ggplot2)
library(ggthemes)
```

```r
df <- mpg
print(head(df))
```

```r
# histogram of hwy mpg values
ggplot(df, aes(x = hwy)) + geom_histogram(binwidth = 2, fill = "red", alpha = 0.5)
```

```r
# barplot of car counts per manufacturer with color fill defined by cyl count
box <- ggplot(df, aes(x = manufacturer)) + geom_bar(aes(fill = factor(cyl))) + theme_gdocs()
print(box)
```

```r
df <- txhousing
print(head(df))
```

```r
# create a scatterplot of volume versus sales
scat <- ggplot(data = df, aes(x = sales, y = volume)) + geom_point(color = "blue", alpha = 0.5)
print(scat)
```

```r
scat <- ggplot(data = df, aes(sales, volume)) + 
geom_point(color = "blue") + 
geom_smooth(color = "red")
print(scat)
```
