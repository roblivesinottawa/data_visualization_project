```r
library(ggplot2)
library(ggthemes)
```

```r
df <- mpg
print(head(df))
```

```r
# histogram of hwy mpg values
ggplot(df, aes(x = hwy)) + geom_histogram(binwidth = 2, color = "pink", fill = "pink")
```

```r
# barplot of car counts per manufacturer with color fill defined by cyl count
box <- ggplot(df, aes(manufacturer)) + geom_bar(aes(fill = cyl), position = "dodge")
print(box)
```

```r
df <- txhousing
print(head(df))
```

```r
# create a scatterplot of volume versus sales
scat <- ggplot(data = df, aes(sales, volume)) + geom_point(color = "blue")
print(scat)
```

```r
scat <- ggplot(data = df, aes(sales, volume)) + geom_point(color = "blue") + geom_smooth(method = "lm", se = TRUE, color = "red")
print(scat)
```
