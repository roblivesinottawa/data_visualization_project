"FEATURE ENGINEERING -> ADD THREE MORE STATISTICS: BATTING AVERAGE, ON-BASE PERCENTAGE, AND SLUGGING PERCENTAGE"

```r
df$BA <- df$H / df$AB
```

```r
print(tail(df$BA, 5))
```

```r
# on base percentage
df$OBP <- (df$H + df$BB + df$HBP) / (df$AB + df$BB + df$HBP + df$SF)
```

```r
# X1B (singles)
df$X1B <- df$H - df$X2B - df$X3B - df$HR
```

```r
# slugging average
df$SLG <- ((1 * df$X1B) + (2 * df$X2B) + (3 * df$X3B) + (4 * df$HR)) / df$AB
```
